```python
# Imported libraries for Snowflake connection
import json
from snowflake.snowpark import Session  # to create a session to connect to Snowflake

def get_session():                      # function to create Snowflake session
    connection_config = {
        "role": "PUBLIC",
        "warehouse": "TRAINING_ROLE FERRET_WH",
        "database": "FERRET_DB",
        "schema": "PUBLIC"
    }
    return Session.builder.configs(connection_config).create()

session = get_session()
print(f"Connected to Snowflake: {session.get_current_account()}")
```
